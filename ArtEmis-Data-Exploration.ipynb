# ABSTRACT ART

```python
# ChatGPT generated and human verified forms of abstract art
abstract_art_styles = ['Abstract_Expressionism', 'Action_painting', 'Analytical_Cubism',
 'Color_Field_Painting', 'Cubism', 'Minimalism', 'Synthetic_Cubism']

abs_exp = df.query("art_style == 'Abstract_Expressionism'")
act_pai = df.query("art_style == 'Action_painting'")
ana_cub = df.query("art_style == 'Analytical_Cubism'")
col_fil = df.query("art_style == 'Color_Field_Painting'")
cubism = df.query("art_style == 'Cubism'")
mini = df.query("art_style == 'Minimalism'")
syn_cub = df.query("art_style == 'Synthetic_Cubism'")

#print(abs_exp.head())
#print(act_pai.head())
#print(ana_cub.head())
#print(col_fil.head())
#print(cubism.head())
#print(mini.head())
#print(syn_cub.head())
```

```python
abstract_art_df = pd.concat([abs_exp, act_pai, ana_cub, col_fil, cubism, mini, syn_cub])

print("================\nArt Style Counts\n================")
print(abstract_art_df['art_style'].value_counts())
print("\n================\nEmotion Description Counts\n================")
print(abstract_art_df['emotion'].value_counts())
print("\n================\nDescriptions per Painting\n================")
print(abstract_art_df['painting'].value_counts())
```

# CONCRETE ART

```python
concrete_art_styles = ['Art_Nouveau_Modern', 'Baroque', 'Contemporary_Realism', 'Early_Renaissance',
                       'Expressionism', 'Fauvism', 'High_Renaissance', 'Impressionism', 'Mannerism_Late_Renaissance',
                       'Naive_Art_Primitivism', 'New_Realism', 'Northern_Renaissance', 'Pointillism', 'Pop_Art',
                       'Post_Impressionism', 'Realism', 'Rococo', 'Romanticism', 'Symbolism', 'Ukiyo_e']

art_nov = df.query("art_style == 'Art_Nouveau_Modern'")
bar = df.query("art_style == 'Baroque'")
con_rea = df.query("art_style == 'Contemporary_Realism'")
ear_ren = df.query("art_style == 'Early_Renaissance'")
exp = df.query("art_style == 'Expressionism'")
fau = df.query("art_style == 'Fauvism'")
high_ren = df.query("art_style == 'High_Renaissance'")
imp = df.query("art_style == 'Impressionism'")
man_lat = df.query("art_style == 'Mannerism_Late_Renaissance'")
nai_art = df.query("art_style == 'Naive_Art_Primitivism'")
new_rea = df.query("art_style == 'New_Realism'")
nor_ren = df.query("art_style == 'Northern_Renaissance'")
poi = df.query("art_style == 'Pointillism'")
pop_art = df.query("art_style == 'Pop_Art'")
post_imp = df.query("art_style == 'Post_Impressionism'")
rea = df.query("art_style == 'Realism'")
roc = df.query("art_style == 'Rococo'")
rom = df.query("art_style == 'Romanticism'")
sym = df.query("art_style == 'Symbolism'")
uki = df.query("art_style == 'Ukiyo_e'")

concrete_art_df = pd.concat([art_nov, bar, con_rea, ear_ren, exp, fau, high_ren,
                             imp, man_lat, nai_art, new_rea, nor_ren, poi, pop_art, post_imp, rea, roc, rom, sym, uki])

print("================\nArt Style Counts\n================")
print(concrete_art_df['art_style'].value_counts())
print("\n================\nEmotion Description Counts\n================")
print(concrete_art_df['emotion'].value_counts())
print("\n================\nDescriptions per Painting\n================")
print(concrete_art_df['painting'].value_counts())
```

# STATS FOR BOTH ABSTRACT AND CONCRETE

```python
abs_count = len(abstract_art_df)
con_count = len(concrete_art_df)

abs_SE_count = len(abstract_art_df[abstract_art_df['emotion'] == 'something else'])
con_SE_count = len(concrete_art_df[concrete_art_df['emotion'] == 'something else'])

print(f"abstract \"something else\" classification percent: {abs_SE_count}/{abs_count} = {(abs_SE_count / abs_count):.2%}")
print(f"concrete \"something else\" classification percent: {con_SE_count}/{con_count} = {(con_SE_count / con_count):.2%}")
```

```python
#This reveals questions about Cubism and Pop Art
SE_percents = dict()

idx = 0

print("=====================\nAbstract Art Styles\n=====================")
for style in (abstract_art_styles + concrete_art_styles):
  print(f"{style:<28} \"something else\" classification percent: {len(df[(df['emotion'] == 'something else') & (df['art_style'] == style)]):>4}/{len(df[df['art_style'] == style]):<5} \
  = {(len(df[(df['emotion'] == 'something else') & (df['art_style'] == style)]) / len(df[df['art_style'] == style])):.2%}")

  SE_percents[style] = len(df[(df['emotion'] == 'something else') & (df['art_style'] == style)]) / len(df[df['art_style'] == style])

  idx += 1
  if idx == 7:
    print("\n=====================\nConcrete Art Styles\n=====================")
```

```python
idx = 0

print("=====================\nAbstract Art Styles\n=====================")
for style in sorted(SE_percents.items(), key=lambda item: item[1], reverse=True):
  print(f"{style[0]:<28} \"something else\" classification percent: {style[1]:.2%}")

  idx += 1
  if idx == 7:
    print("\n=====================\nConcrete Art Styles\n=====================")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

desc_per_abs_paint = abstract_art_df['painting'].value_counts().value_counts().sort_index()
desc_per_all_paint = df['painting'].value_counts().value_counts().sort_index()

# Use the x-values as the actual number of descriptions
x = desc_per_all_paint.index

# Bar width (adjustable)
width = 0.45

# Create figure
plt.figure(figsize=(15,9))

# Plot side-by-side bars using numeric positions
plt.bar(x - width/2, desc_per_all_paint.reindex(x, fill_value=0), width=width, color='#E67E22', label='All Art')
plt.bar(x + width/2, desc_per_abs_paint.reindex(x, fill_value=0), width=width, color='#3498DB', label='Abstract Art')

# Labels and title
plt.xlabel("# of Descriptions", fontsize=21)
plt.ylabel("# of Paintings with that # of Descriptions", fontsize=21)
plt.title("Distribution of Descriptions per Painting", fontsize=24)

# X-axis spacing: show all numeric labels, rotated for readability
plt.xticks(x, fontsize=11)

# Log scale on y-axis
plt.yscale('log')

# Gridlines for readability
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Legend
plt.legend(title="Art Type", fontsize=18, title_fontsize=18)

# Tight layout for proper spacing
plt.tight_layout()

plt.show()
```
