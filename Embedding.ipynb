```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import torch

# Check for GPU availability and set device accordingly
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")
```

### Load Files and Model

```python
# Load Images CSV
file_path = f"/content/drive/Shareddrives/LLMs_Art_Project/Data/Results/Llama/llama_results_0shot.csv"
df = pd.read_csv(file_path)
```

```python
# Load the BGE-Base model, using the determined device
model = SentenceTransformer("BAAI/bge-base-en-v1.5", device=device)
```

### Creating vectors for each utterance

```python
# Extract list of utterances
texts = df["generated_description"].tolist()

# Batch encode all at once
embs = model.encode(
    texts,
    batch_size=32,
    convert_to_tensor=False,
    show_progress_bar=True
)

# Save as lists
df["embeddings"] = [e.tolist() for e in embs]
```

```python
# Save to CSV
df.to_csv("/content/drive/Shareddrives/LLMs_Art_Project/Data/Results/Results_With_Embeddings/llama_results_0shot_embedded.csv", index=False)
```
